**Public and Non Public Attributes**

**Attributes: ** There are 2 type of Attributes.
1. **Public Attribute:** A Public attribute is an attribute that can be accessed and modified directly without access restrictions.
2. **Non Public Attribute:** A Non public attribute is an attribute that should not be accessed or modified  outside of the class.

In python there are 2 ways of defining Non Public attribute:
1. By Convention: We need to add the _(attribute Name) eg _Name
2. Changing Name( More difficult to access) __(attribute_name). it is difficult to access as we are changing the name while access the attribute. This is also called **Name Mangling**. this should on be used for special case.

```python
class Movie:

  def __init__(self , title, rating):
    self.title = title
    self._rating =rating

  @property
  def rating(self):
    return self._rating
  @rating.setter
  def rating(self, new_rating):
    if 1.0 <= new_rating <= 5.0 and isinstance(new_rating, float):
      self._rating= new_rating
    else:
      print("Please enter a valid rating")


fav_movie = Movie("Titanic", 4.3)
print(fav_movie.rating)
fav_movie.rating=4.5
print(fav_movie.rating)


```

```python
class Backpack:
  def __init__(self):
    self._items = []

  @property
  def items(self):
    return self._items

  @items.setter
  def items(self, new_items):
    if isinstance(new_items, list):
      self._items=new_items
    else:
      print("Enter a list of element that need to be added in backpack")

my_backpack = Backpack()
print(my_backpack.items)
my_backpack.items=["Books", "Pen"]
print(my_backpack.items)
```

```python
from types import EllipsisType
# Assingment 6

class BouncyBall:
    def __init__(self, price, size, brand):
        self._price = price
        self._size = size
        self._brand = brand

    @property
    def price(self):
        return self._price
    @price.setter
    def price(self, new_price):
      if isinstance(new_price, float) and new_price>0:
        self._price = new_price
      else:
        print("Enter valid price")


    @property
    def size(self):
        return self._size
    @size.setter
    def size(self, new_size):
      if isinstance(new_size, int) and new_size>0:
        self._size=new_size
      else:
        print("Enter valid size")

    @property
    def brand(self):
        return self._brand
    @brand.setter
    def brand(self, new_brand):
      if isinstance(new_brand, str):
        self._brand=new_brand
      else:
        print("Enter valid brand")


ball = BouncyBall(10, "Small", "Nike")

print(ball.price)
print(ball.size)
print(ball.brand)

ball.brand ="xyz"
print(ball.price)
print(ball.size)
print(ball.brand)


```
