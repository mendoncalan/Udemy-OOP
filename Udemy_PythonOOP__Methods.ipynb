# **Method**
A function associated to an object of the class or to the class itself is called a method.

The Methods defined in a class determine the behavior of the objects created from the class and how they can interact with their state.

A method is a function that belongs to an object,

Calling a method is similar to calling a function. here we are asking the code to execute it functionality.

Method names usually include verbs since they represents actions.

```python
class Backpack:
  def __init__(self):
    self._items = []

  @property
  def items(self):
    return self._items
  def add_multiple_items(self, items):
    for item in items:
      self.add_item(item) # calling the method from another method

  def add_item(self,  item):
    if isinstance(item, str):
      self._items.append(item)
    else:
      print("Please enter the valid method")
  def remove_item(self, item):
    if isinstance(item, str) and item in self._items:
      self._items.remove(item)
      print(f"Item was remove {item}")
    else:
      print("Enter is not present in the list so could not be removed")
  def has_item(self, item):
    if isinstance(item, str) and item in self._items:
      print("Item is present in the list")
    else:
      print("Item is not present in the list")

  def show_items(self, sorted_list=False):
    if sorted_list:
      print(sorted(self._items))
    else:
      print(self._items)

my_Backpack =Backpack()
my_Backpack.add_item("Water Bottle")
my_Backpack.add_item("Sleeping Bag")
my_Backpack.add_item("Bottle")
my_Backpack.add_item("Cup")
my_Backpack.add_item("Book")
my_Backpack.add_multiple_items(["Water Botter", "Canady"])
print(my_Backpack.items)
my_Backpack.show_items()
my_Backpack.show_items(True)


# Calling Methods from other methods
```
