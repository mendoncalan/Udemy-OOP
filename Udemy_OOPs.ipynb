Instance Attribute and __init__

>Classes define the states and the behavior of the objects in a generic way


1.   __init__() method: Special method uusd to define the initial state of an object.it is called automatically when an instance is created.


*   Parameter : There values that passed with the () paranteises are called parameter  price, color, bedroom are parameter.
*   Attribute: the variable that take the value of parameter are attribute/ it is the property of the class. eg self.color


2.   self : is a generic way of referring to the current instance of the class. self represents the instance that is being created.  

3. Instance creation: self is skipped in the arguments list when we create an instance.

4. Default Arguments: Default values assigned to parameters when their corresponding arguments are omitted. The default paramenter should be last paramenter in the list of parametes.


```python
class House:
  def __init__(self,price ,color,bedrooms):
    self.price = price # here price is a parameter and self.price is the attribute.
    self.color = color
```

```python
class Backpack:
  def __init__(self, color='Black', size=0):# setting Default Arguments
    self.color = color
    self.size = size
    self.items = []
    print(self.items)
  def add_item(self,item):
    self.items.append(item)


# Creating an instance of the Class
my_backpack = Backpack("Blue", "Medium")
my_backpack.add_item("Laptop")
my_backpack.add_item("Phone")

#Accessing the instance of Attributes.
print(my_backpack.items)
```

```python
class Circle:
  def __init__(self, radius):
    self.radius = radius
    self.color = "Blue"

class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width
    self.color = "Red"
class Movie:
  def __init__(self, title, year, language,  rating):
    self.title = title
    self.year = year
    self.language = language
    self.rating = rating

# Creating an instance of the Class
my_circle = Circle(5)
my_rectangle = Rectangle(10, 20)
my_movie = Movie("The Matrix", 1999, "English", "PG-13")

#
#Accessing the instance of Attributes.
print(my_movie.language)
print(my_movie.rating)
print(my_circle.color)
print(my_rectangle.color )
```

```python

class Backpack:
  def __init__(self, color='Black', size=0):# setting Default Arguments def __init__(self, size,color='Black')
  #def __init__(self, color, size=5) :default values of the arguments should be defined at the last of in init method.
    self.color = color
    self.size = size
    self.items = []
    print(self.items)
  def add_item(self,item):
    self.items.append(item)


# Creating an instance of the Class
my_backpack = Backpack("Blue", "Medium")
my_backpack.add_item("Laptop")
my_backpack.add_item("Phone")

#Accessing the instance of Attributes.
print(my_backpack.items)

# Moddify the value of instance Attributes.
print(my_backpack.color)
my_backpack.color= 'Red'
print(my_backpack.color)
```

```python
#Iterating Over Sequences of Objects
#for Loops and Instances
class Player:

    def __init__(self, x, y):
        self.x = x
        self.y = y
player1 = Player(5, 6)
player2 = Player(2, 4)
player3 = Player(3, 6)


players = [player1, player2, player3]
for player in players:
    print(f"X: {player.x} Y: {player.y}")
```
