Instance Attribute and __init__

>Classes define the states and the behavior of the objects in a generic way


1.   __init__() method: Special method uusd to define the initial state of an object.it is called automatically when an instance is created.


*   Parameter : There values that passed with the () paranteises are called parameter  price, color, bedroom are parameter.
*   Attribute: the variable that take the value of parameter are attribute/ it is the property of the class. eg self.color


2.   self : is a generic way of referring to the current instance of the class. self represents the instance that is being created.  

3. Instance creation: self is skipped in the arguments list when we create an instance.

4. Default Arguments: Default values assigned to parameters when their corresponding arguments are omitted. The default paramenter should be last paramenter in the list of parametes.

**How to Delete an instance Attribute**
There are two ways to do this in Python.

*    **Delete Instance Attributes with del** : To delete a specific instance attribute, you can use the del keyword followed by a space, the name of the instance, a dot, and the name of the attribute that will be deleted.

```
syntax : del instance.attribute

```


*   **Delete Instance Attributes with delattr** : Alternatively, you can use delattr (a built-in Python function) to delete attributes dynamically based on the value of a variable.


```
Syntax: delattr(instance, attribute)
```




````python
class Player:

    def __init__(self, x, y):
        self.x = x
        self.y = y
player = Player(6, 8)
attributes = ['x', 'y']
for attribute in attributes:
    print(player, attribute)
    delattr(player, attribute)
#print(player.x)
#print(player.y)
````

**Class Attributes**
Class Attribute is an attribute of the class. They belong to class.


*   All the instances of the class have access to this attribute
*   They share the same value, so any changes made to this values affects all instances.



```
Syntax:

class ClassName:
  # Class Attributes
  #__init__()
  # Methods
```

**Class Attributes vs. Instance Attributes**



*   **Class Attributes**



1.  They belong to the class itself.
2.  Changing their value affects all the instances of the class because they take the value from the same source.




*   **Instance Attributes**


1.   They belong to the instances.
2.   Every instance has its own, independent, copy of the attribute.
3.   Changing their value only affects a particular instance. Other instances are not modified.



**When to Use Class Attributes**

**Defining Constants** :A common use case is storing constants that are very closely related to the class.


```
class Earth:
    MEAN_RADIUS = 6371
    EQUATORIAL_RADIUS = 6378.1
    EQUATORIAL_CIRCUMFERENCE = 40075.017
```
**Setting Default Values** : You can also use class attribute if you need to set a default value for all the instances of the class.

```
class IceCream:

    discount = 15

    def __init__(self, discount=None):
        if discount is not None:
            self.discount = discount
```

**Counting Instances**
If you need to know how many instances of a class have been created, you can use a class attribute to keep track of the counter.


```
class Customer:

    customer_id = 0

    def __init__(self, name):
        self.name = name
        self.id = Customer.customer_id
        Customer.customer_id += 1
```
